```python
import sys, os
sys.path.insert(0, os.path.abspath("/home/tom/tesi/MCSHeuristicCpp"))

import argparse

import torch

import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 16})

from common import models
from common import utils
from subgraph_matching.config import parse_encoder
import analyze_sim
import analyze_results.test_utility as test_utility
import random
```

## Load model

```python
parser = argparse.ArgumentParser()

utils.parse_optimizer(parser)
parse_encoder(parser)
args = parser.parse_args("")
args.model_path = os.path.join("..", "ckpt/model.pt")

print("Using dataset {}".format(args.dataset))
model = models.OrderEmbedder(1, args.hidden_dim, args)
model.to(utils.get_device())
model.eval()
model.load_state_dict(torch.load("/home/tom/tesi/MCSHeuristicCpp/ckpt/model.pt", map_location=utils.get_device()))
```

```python
mc = []
for filename in sorted(os.listdir("/home/tom/tesi/MCSHeuristicCpp/test_results/classic/")):
    tt = test_utility.load_test("/home/tom/tesi/MCSHeuristicCpp/test_results/classic/" + filename)
    mc.append(tt)

tot_avg_sol = 0
tot_avg_rand = 0
for mct in mc:
    print(mct.test_description[0])
    graphs = []
    for graphPath in sorted(os.scandir("/home/tom/tesi/MCSHeuristicCpp/text_ben/fast_test/" + mct.test_description[0]), key=lambda e: e.name):
                        graphs.append(graphPath)

    sim_dict, n = analyze_sim.launch_solver(input_files=graphs, vertex_labeled=False, directed=False,
                connected=False, edge_labeled=False, timeout=3000, model=model, heuristic = 'classic', hops=2, batch_size=1)

    random_sol = []
    for i in range(len(mct.solution)):
        random_sol.append((random.randint(0, n-1), random.randint(0, n-1)))

    avg_sol = 0
    avg_rand = 0
    for pair in mct.solution:
        avg_sol += sim_dict[(pair[0], pair[1])]
        # print(sim_dict[(pair[0], pair[1])])
    for pair in random_sol:
        avg_rand += sim_dict[(pair[0], pair[1])]
    avg_sol /= len(mct.solution)
    avg_rand /= len(mct.solution)

    print(avg_sol)
    print(avg_rand)
    print("")

    tot_avg_sol += avg_sol
    tot_avg_rand += avg_rand 

tot_avg_sol /= len(mc)
tot_avg_rand /= len(mc)
print("TOTAL")
print(tot_avg_sol)
print(tot_avg_rand)

```
